```python
# Imports
import pandas as pd
import openpyxl
import os
```

```python
# Import do Dataset Retail rocket

# Carregando os datasets
df_category = pd.read_csv("../data/raw/retailrocket/category_tree.csv")
df_events = pd.read_csv("../data/raw/retailrocket/events.csv")
df_item_part1 = pd.read_csv("../data/raw/retailrocket/item_properties_part1.csv")
df_item_part2 = pd.read_csv("../data/raw/retailrocket/item_properties_part2.csv")
```

```python
# análise do dataset category_tree
print(df_category.head())
print(df_category.info())
print(df_category.describe())
```

Dataset com tabela de relacionamento entre id de categoria de produto e id parent

```python
# análise do dataset events
print(df_events.head())
print(df_events.info())
print(df_events.describe())

#salvando amostra aleatória do dataset events em Excel
df_events_sample = df_events.sample(n=10000, random_state=42)
os.makedirs("../data/interim", exist_ok=True)
df_events_sample.to_excel("../data/interim/events_sample.xlsx", index=False)
```

```python
print(df_events[df_events['transactionid'].notna()].head())
# ver distribuição de eventos com e sem transactionid

print(df_events[df_events['transactionid'].isna()].count())
print(df_events[df_events['transactionid'].notna()].count())
```

```python
# análise dos datasets item_properties_part1 e item_properties_part2
print(df_item_part1.head())
print(df_item_part1.info())
print(df_item_part1.describe())

print(df_item_part2.head())
print(df_item_part2.info())
print(df_item_part2.describe())

# Salvando amostra aleatória única com união de amostras dos datasets 
# item_properties_part1 e item_properties_part2 em Excel
df_item_sample = (pd.concat([df_item_part1, df_item_part2])
                  .sample(n=30000, random_state=42))
os.makedirs("../data/interim", exist_ok=True)
df_item_sample.to_excel("../data/interim/item_properties_sample.xlsx", index=False)
```
